```python
import matplotlib as mplt
import plotly as plt
import tensorflow as tf
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_olivetti_faces
```

```python
dataset = fetch_olivetti_faces()
```

```python
X = dataset['data']
Y = dataset['target']
images = dataset['images']
```

```python
%matplotlib inline
```

```python
fig = mplt.pyplot.figure()
for i in range(30): 
    img_grid = fig.add_subplot(5, 6, i+1)
    img_grid.imshow(images[i])

mplt.pyplot.show()
```

```python
pd.DataFrame(X)
```

```python
pca = PCA(n_components=40)
```

```python
pca.fit(X)
```

```python
mplt.pyplot. plot(pca.explained_variance_ratio_)
```

```python
X_new = pca.fit_transform(X)
```

```python
color = sns.color_palette(None, 40)
```

```python
color
```

```python
X_new = np.array(X_new)
Y = np.array(Y)
```

```python
Y.shape, X_new.shape
```

```python
colorsDataset = []
for i in range(400):
    colorsDataset.append(color[Y[i]])
```

```python
colorsDataset[0]
```

```python
mplt.pyplot.scatter(X_new[:,0], X_new[:,1],color = colorsDataset )
```

```python
import plotly.express as px
```

```python
fig = px.scatter_3d(x= X_new[:,0], y= X_new[:,1], z=X_new[:,2], color=colorsDataset)
fig.update_layout(showlegend=False)
fig.show(renderer="png")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_new, Y, test_size=0.5, random_state=0)
```

```python
gnb = GaussianNB()
y_pred = gnb.fit(X_train, y_train).predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, plot_confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
table = pd.DataFrame(confusion_matrix(y_test,y_pred))
```

```python
pd.set_option('display.max_columns',table.shape[0]+1)
table
```
